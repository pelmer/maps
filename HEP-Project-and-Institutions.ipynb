# Create a map of US Institutions receiving funding from HEP Ops programs and/or Institute class projects
#

```python
import folium
from folium.features import DivIcon
import pandas as pd
```

```python
df = pd.read_csv('US-HEP-Institutions-Ops-and-Institute-Funding.csv')

m = folium.Map(location=[40.0, -95.0], zoom_start=4)
for i, row in df.iterrows():
    latitude = row['Latitude']
    longitude = row['Longitude']
    tooltip = row['Institution']
    popup = row['Institution'] + ":\n"
    project = False
    if row['USATLAS Ops']==1:
        popup += "USATLAS Operations (DOE/NSF)\n"
        project = True
    if row['USCMS Ops']==1:
        popup += "USCMS Operations (DOE/NSF)\n"
        project = True
    if row['DUNE Ops']==1:
        popup += "DUNE Operations (DOE)\n"
        project = True
    if row['IRIS-HEP']==1:
        popup += "IRIS-HEP Software Institute (NSF)\n"
        project = True
    if row['A3D3']==1:
        popup += "A3D3 HDR Institute (NSF)\n"
        project = True
    insttype = row['Type']
    mycolor = 'green'   # default for anything other than 2
    if project:
        mycolor = 'blue'
    if insttype==2:   # Just call out the DOE labs
        mycolor = 'red'
    #folium.CircleMarker(location=[latitude, longitude], radius=4, fill=True, fill_color=mycolor, fill_opacity=1.0, color=mycolor, tooltip=tooltip, popup=popup).add_to(m)
    folium.Marker(location=[latitude, longitude], icon=folium.DivIcon(
        icon_size=(150,36),
        icon_anchor=(7,20),
        html=f'<div style="font-size: 18pt; color: {mycolor}">â˜…</div>',
    ),radius=4, fill=True, fill_color=mycolor, fill_opacity=1.0, color=mycolor, tooltip=tooltip, popup=popup).add_to(m)
m.save("map.html")
```
